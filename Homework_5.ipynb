## 1. Manhattan Distance

## 3. Cosine Similarity

In the following, the cosine similarity between the customers will be calculated. The results are hereby visualized in a similarity matrix, where the diagonal corresponds to the similarity between the same customers and is therefore 1.

```python
from numpy import dot
from numpy.linalg import norm
X_CS = df_1

arr_cs = np.zeros((len(X_CS), len(X_CS)))
for i in range(len(X_CS)):
    for j in range(len(X_CS)):
        arr_cs[(i),(j)] = dot(X_CS[i], X_CS[j])/(norm(X_CS[i])*norm(X_CS[j]))
```

```python
df_cs = pd.DataFrame(arr_cs, columns = ID)
df_cs.index = ID
df_cs.head()
```

## 4. Simple Matching

In the following, the simple matching coefficient between the customers will be calculated. The results are hereby visualized in a similarity matrix, where the diagonal corresponds to the similarity between the same customers and is therefore 1.

```python
def simple_matching(x,y):
    x_test = np.logical_not(x).tolist()
    x_n = [int(val) for val in x_test]
    y_test = np.logical_not(y).tolist()
    y_n = [int(val) for val in y_test]
    its = np.logical_and(x, y)
    its_n = np.logical_and(x_n, y_n)
    union = np.logical_or(x, y)
    union_n = np.logical_or(x_n, y_n)
    sim = (its.sum() + its_n.sum()) / (float(union.sum()) + its_n.sum())
    return sim
```

```python
X_SM = df_sm

arr_sm = np.zeros((len(X_SM), len(X_SM)))
for i in range(len(X_SM)):
    for j in range(len(X_SM)):
        arr_sm[(i),(j)] = simple_matching(X_SM[i], X_SM[j])
```

```python
df_smc = pd.DataFrame(arr_sm)
df_smc.head()
```

## 5. Jaccard Similarity

In the following, the Jaccard Similarity between the customers will be calculated. The results are hereby visualized in a similarity matrix, where the diagonal corresponds to the similarity between the same customers and is therefore 1.

```python
X_JS = df_sm

def jaccard_binary(x,y):
    its = np.logical_and(x, y)
    union = np.logical_or(x, y)
    sim = its.sum() / float(union.sum())
    return sim

arr_js = np.zeros((len(X_JS), len(X_JS)))
for i in range(len(X_JS)):
    for j in range(len(X_JS)):
        arr_js[(i),(j)] = jaccard_binary(X_JS[i], X_JS[j])
```

```python
df_js = pd.DataFrame(arr_js)
df_js.head()
```

## 6. Clustering k-Means Algorithm

In the previous calculations, different ways of computing distances and similarities were presented. One way to make use of that is to cluster the data in certain groups. In the following, the k-Mean Algorithm is used and clusters the dataframe into 5 different groups. The algorithm is hereby based on the euclidean distance. 

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=0).fit(df1)
li = kmeans.labels_
li = li.tolist()
df1['Cluster'] = li
df1.head()
```

In the dataframe above, the column "Cluster" gives information about which cluster the customer is assigned to. 
To get more insights about the distribution of the customers in the clusters, the following output gives more information. By that, it get's clear that the customers are not equally distributed, especially in the third cluster. 

```python
unique, counts = np.unique(li, return_counts=True)
print(np.asarray((unique, counts)).T)
```
